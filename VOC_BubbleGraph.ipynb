```python
import pandas as pd

# Load the dataset from the Excel file
data_path = 'Bubble graph test_VOC.xlsx'
voc_data = pd.read_excel(data_path)

# Display the first few rows of the dataset
voc_data.head()
```

```python
  import seaborn as sns
import matplotlib.pyplot as plt
import openpyxl

# Set the figure size
plt.figure(figsize=(12, 8))

# Create the bubble graph using seaborn
sns.scatterplot(data=voc_data, x='Main category (x)', y='satisfaction (y)', size='no of mention (size of bubble)',
                hue='Main category (x)', palette='tab10', sizes=(100, 2000), legend='full')

# Set the title and labels
plt.title('Voice of Customer Bubble Graph')
plt.xlabel('Main Category')
plt.ylabel('Satisfaction')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the graph
plt.show()
```

```python
# Create the bubble graph again
plt.figure(figsize=(12, 8))
sns.scatterplot(data=voc_data, x='Main category (x)', y='satisfaction (y)', size='no of mention (size of bubble)',
                hue='Main category (x)', palette='tab10', sizes=(100, 2000), legend='full')
plt.title('Voice of Customer Bubble Graph')
plt.xlabel('Main Category')
plt.ylabel('Satisfaction')
plt.xticks(rotation=45)

# Add the satisfaction spectrum at the bottom
cmap = plt.get_cmap('RdYlGn')
colors = [cmap(i) for i in [0, 0.5, 1]]
labels = ['Very Dissatisfied', 'Neutral', 'Satisfied']
positions = [0, 0.5, 1]

ax2 = plt.axes([0.2, 0.05, 0.6, 0.03], frameon=False)
ax2.imshow([colors], aspect='auto', cmap='RdYlGn')
ax2.set_xticks(positions)
ax2.set_xticklabels(labels)
ax2.set_yticks([])

plt.tight_layout()
plt.show()
```
